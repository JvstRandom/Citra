```python
import pandas as pd
import plotly.express as px
import numpy as np 
import matplotlib.pyplot as plt

# Read data from CSV
df = pd.read_csv('histogram21s.csv')


# Plotting the grouped bar chart with annotations
bar_width = 0.3
index = np.arange(len(df['Kategori']))


fig, ax = plt.subplots(figsize=(12, 8))


cities = df.columns[1:]


for i, city in enumerate(cities):
    bars = ax.bar(index + i * bar_width, df[city], bar_width, label=city, alpha=0.7)
    for bar in bars:
        yval = bar.get_height()
        ax.text(bar.get_x() + bar.get_width() / 2, yval, round(yval, 1), ha='center', va='bottom', color='black', fontweight='normal')


# Adding faint grid lines
ax.grid(color='gray', linestyle='-', linewidth=0.25, alpha=0.5)


# Adding labels and title
ax.set_xlabel('Kategori')
ax.set_ylabel('Frekuensi')
ax.set_title('Usia')
ax.set_xticks(index + bar_width * (len(cities) - 1) / 2)
ax.set_xticklabels(df['Kategori'])
ax.legend(title='Keterangan')


# Display the plot
plt.show()
```
